# first task 
### write a program that counts up the number of vowels contained in the string

```python
#%%writefile vowels_count_function.py
def count_vowels(string):
    try:
        # Check if the input is a string
        if not isinstance(string, str):
            raise ValueError("Input must be a string.")

        # Define vowels as a set for O(1) lookup
        vowels = {"a", "e", "i", "o", "u"}

        # Convert the string to lowercase and count vowels
        return sum(1 for char in string.lower() if char in vowels)

    except ValueError:
        return "Invalid input"
```

# second task
### write a program that prints the location of "i" char in any string you added 

```python
def find_i(string):
    index=0
    for char in string:
        if char.lower()=="i":
            return print(f"The character 'i' is found at index {index}")
    
        index+=1
    print("'i' character isn't existed")
```

```python
find_i("the world s good")
```

# another way

```python
#%%writefile find_index_of_i.py
def find_i(string):
    return [index for index,char in enumerate(string) if char =="i"]
```

```python
find_i("this is the will we will")
```

# fourth task 
### mario pyramid

```python
#%%writefile mariopyramid.py
def mario_pyramid(rows):
    for i in range(1,rows+1):
        print(" " *(rows-i) ,"*"*i)
        
        
```

```python
def get_numbers_from_user():
    numbers=[]
    max_invalid_attempts = 5 
    for i in range(5):
        invalid_attempts = 0
        while True:
            try:
                num=int(input(f"please enter number {i+1} of 5 "))
                numbers.append(num)
                break
            except ValueError:
                invalid_attempts+=1
                print("Invalid input, please enter a valid number. ")
                if invalid_attempts>max_invalid_attempts:
                    raise("you've reached the maximum tries")
        
                
                
    return numbers
        
def sort_list():
    numbers=get_numbers_from_user()
    ascending_numbers = sorted(numbers)
    descending_numbers = sorted(numbers, reverse=True)

    print("\nSorting Results:")
    print(f"Ascending order:  {ascending_numbers}")
    print(f"Descending order: {descending_numbers}")
```

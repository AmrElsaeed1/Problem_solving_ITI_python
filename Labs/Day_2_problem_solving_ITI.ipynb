# Task one 
### fill an array of 5 elements from the user, fort it in descending and ascending orders, then display the output

```python
number=[]
for i in range(0,5):
    num=input("please input a number: ")
    while not num.isdigit():
        num=input("please input a valid number: ")
        
    else:
        number.append(int(num))
        
acending_number=sorted(number)
descending_list=sorted(number,reverse=True)
print(f"the ascending list is {acending_number}")
print(f"the descending list is {descending_list}")
```

# task two 
### write a program that generate a multiplication table from 1 to the number passed

```python
num = input("Enter a number: ")
while not num.isdigit():
    num=input("please input a valid number: ")
mul=[]
num=int(num)
for i in range(1, num + 1):
    temp=[]
    for j in range(1, 13):
        if j > i:
            break
        product = i * j
        temp.append(product)
        
    mul.append(temp)
        
# mul.append(product)
print(mul)
# you can make del temp ,because it's help variable ,not essensial
```

# bouns

```python
while True:
    Email = input("Please enter your email: ").strip()  
    
    if "@" not in Email or "." not in Email:
        print("Invalid email")
        continue
    
    at_index = Email.find("@")
    dot_index = Email.find(".")
    
    if at_index == 0 or at_index == len(Email) - 1:
        print("Invalid email")
        continue
    
    if dot_index == 0 or dot_index == len(Email) - 1:
        print("Invalid email")
        continue
    
    if dot_index < at_index:
        print("Invalid email")
        continue
    if abs(dot_index-at_index)==1:
        print("Invalid email")
        continue
        
    
    print("Valid email!")
    break
```

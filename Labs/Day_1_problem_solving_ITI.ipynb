# first task 
### write a program that counts up the number of vowels contained in the string

```python
count=0
string="The world is good"
vowels=["a","e","i","u","o"]
for char in string:
    if char.lower() in vowels:
        count+=1
print(f"the Total number of vowels chars is : {count}")
```

# third task
### write a program that generate multiplication table from 1 to the number passed

```python

num = int(input("Enter a number: "))

for i in range(1, num + 1):
    for j in range(1, 13):
        if j > i:
            break
        product = i * j
        print(f"{i} x {j} = {product}")
```

# fourth task 
### mario pyramid

```python
rows = 5 
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * i)
```
